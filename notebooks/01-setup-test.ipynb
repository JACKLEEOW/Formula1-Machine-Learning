```python
session = fastf1.get_session(2023, 'Monaco', 'R')

session.load()
```

```python
laps = session.laps

print (laps.columns)

print (laps.head())
```

```python
hamilton = laps[laps['Driver'] == 'HAM']

hamilton



```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# .dt.total_seconds() is a pandas helper that converts Timedelta objects to numbers

hamilton['LapTimeSeconds'] = hamilton['LapTime'].dt.total_seconds()

# creating the plot,  first is size
plt.figure(figsize=(10,6)) 

sns.scatterplot(data = hamilton, x = 'LapNumber', y= 'LapTimeSeconds')

#label axis

plt.title("Lewis Hamilton - Monaco 2023 Race Pace")

plt.xlabel("Lap Number")
plt.ylabel("Lap Time (Seconds)")

#Since doing racing, we must invert the Y axis since by default, higher number are at top,
# but in racing the lower number is better

plt.gca().invert_yaxis()

plt.show()

#in the graph we can tell around lap 52, it started to RAIN
# causing hamilton's lap time to get worse

```

```python
# To build a proper model first, we should train only on the DRY laps, 


dry_laps = hamilton[hamilton['LapNumber'] < 50]

# now replot

plt.figure(figsize= (10,6))
sns.scatterplot(data = dry_laps, x = 'LapNumber', y = 'LapTimeSeconds')
plt.title ("Lewis Hamilton Dry Pace Only")
plt.gca().invert_yaxis()
plt.show()
```

```python
# INCOMING MACHINE . LEARN()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Select variables

x = dry_laps[['LapNumber']] #remember X has to be a 2d Matrix
y = dry_laps['LapTimeSeconds'] # Y can be a series/1d vector

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)

print(f"Training Test Shapes : {x_train.shape}")
print(f"Testing shapes: {x_test.shape}")
```

```python
# initialize the ML model to use
model = LinearRegression()
#Teach model to fit the line to training data
model.fit(x_train,y_train)


# Model should now learn the math of tire degradation

print(f"Slope (seconds gained per lap): {model.coef_[0]:.4f}")
print (f"Intercept (Theoretical Lap Time): {model.intercept_:.4f}")
```

```python
# ask model, what should the lap time be for these laps
predictions = model.predict(x)

# plot the actual data
plt.figure(figsize=(10,6))


sns.scatterplot(x = dry_laps['LapNumber'], y=dry_laps['LapTimeSeconds'], label = 'Actual Laps', color = 'blue', alpha = 0.6 )

#plot the model's data
# plot the lap number against our predictions

plt.plot(dry_laps['LapNumber'], predictions, color = 'red', linewidth = 3, label = 'ML Trend Line')

plt.title("Lewis Hamilton - Monaco 2023 AI Pace Predictions")
plt.xlabel("Lap Number")
plt.ylabel("Lap Time (Seconds)")
plt.legend()
plt.gca().invert_yaxis()
plt.show()
```

```python
# Our prior issues was that we did not filter out the time for pit stops
# the outliers are dragging the entire line down.
# so we must clean the data and reinitialize the laps only < 80 seconds


clean_laps = hamilton[(hamilton["LapNumber"] < 50) & (hamilton['LapTimeSeconds'] < 81)]

x_clean = clean_laps[['LapNumber']] #2d Matrix as req

y_clean = clean_laps ['LapTimeSeconds']

#Re train model
print("Shape of x_clean:", x_clean.shape)
print("Shape of y_clean:", y_clean.shape)
model_clean = LinearRegression()
model_clean.fit(x_clean,y_clean)

#plot new results
predictions_clean = model_clean.predict(x_clean)

plt.figure(figsize=(10,6))
sns.scatterplot(data = clean_laps, x = "LapNumber", y = "LapTimeSeconds", label = "Clean Racing Laps (No Pit)", color = "blue")
plt.plot(clean_laps['LapNumber'], predictions_clean, color = "green", linewidth = 3, label = "Machine Prediction")

plt.title("Lewis Hamilton True Pace - No Pit")
plt.legend()
plt.gca().invert_yaxis()
plt.show()  

print(f"Slope (Seconds gained per lap): {model_clean.coef_[0]:.4f}")
print(f" Theoretical Lap time : {model_clean.intercept_:.4f}")
```

```python
#If slope is negative, that means tire wear < fuel effect Hamilton is GAINING Time as race goes on
# if slope is positive that means tire wear > fuel effect Hamilton is LOSING Time as race goes on
```
